```python
from abc import ABCMeta, abstractmethod
```

```python
class PopulationClass:
    
    __metaclass__ = ABCMeta
    
    t = 0
    
    def __init__(self, i, j):
        self.i = i
        self.j = j
    
    @abstractmethod
    def height(self):
        pass
    
    @abstractmethod
    def recource_capacity(self):
        pass
    
    @abstractmethod
    def f(self):
        pass
    
    @abstractmethod
    def g(self):
        pass
    
    @abstractmethod
    def resource_decrease(self):
        pass
    
    def max_capacity(self):
        return self.f(self.height())
        
    def energy_pulse(self, *args):
        pass
        
    def speed_height(self, t):
        return self.g(self.energy_pulse(), self.height())
    
    def resourse_volume(self, t):
        return self.resourse_volume(self, t - 1) + self.speed_height() - self.resource_decrease()
```
